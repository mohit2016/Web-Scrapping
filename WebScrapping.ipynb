```python
import requests
```

```python
import bs4
```

```python
res = requests.get("https://www.analyticsvidhya.com/glossary-of-common-statistics-and-machine-learning-terms")
```

```python
res
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(res.text)
```

```python
# soup variable contains the complete html code for the web page

```

```python
ml_skills = []

strong_skills = soup.select("strong")
b_skills = soup.select("b")

for i in strong_skills:
    print(i.getText())
    ml_skills.append(i.getText())
    
for i in b_skills:
    print(i.getText())
    ml_skills.append(i.getText())
```

```python
len(ml_skills)
```

```python
import json

json.dump(ml_skills, open("machine_learning.json" , "w" , encoding="utf16"))
```

```python
res_web_2 = requests.get("https://www.keycdn.com/blog/web-development-tools")
```

```python
res_web_2
```

```python
soup_web_2 = BeautifulSoup(res_web_2.text)
```

```python
# soup_web_2
```

```python
web_sills_2 = soup_web_2.select("li")
```

```python
web_skl = []
for count, i in enumerate(web_sills_2):
    print(count , i.getText())
    web_skl.append(i.getText())
```

```python
# use only 13 to 152 
web_skl = web_skl[13:152]
```

```python
# cleaning the junk text
for i in web_skl:
    print(i.split(":")[0].strip("/r"))
    web_all_skills.append(i.split(":")[0].strip("/r"))
```

```python
len(web_all_skills)
```

```python
json.dump(web_all_skills,open("web.json", "w", encoding="utf16"))
```
